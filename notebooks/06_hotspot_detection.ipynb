```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Area": ["Zone A", "Zone B", "Zone C", "Zone D", "Zone E"],
    "Predicted_LST": [28.5, 33.2, 37.8, 42.1, 39.5]
}
df = pd.DataFrame(data)

df
```

```python
def classify_heat(lst):

    if lst < 30:
        return "Low Risk"

    elif lst < 35:
        return "Moderate Risk"

    elif lst < 40:
        return "High Risk"

    else:
        return "Extreme Hotspot"
```

```python
df["Heat_Risk"] = df["Predicted_LST"].apply(classify_heat)

df
```

```python
df["Heat_Risk"].value_counts()
```

```python
extreme_hotspots = df[
    df["Heat_Risk"] == "Extreme Hotspot"
]

extreme_hotspots
```

```python
extreme_hotspots.to_csv(
    "../outputs/reports/extreme_hotspots.csv",
    index=False
)

print("Hotspots Saved Successfully")
```
