```python
def audit_data(df):

    print("=" * 50)
    print("DATASET SHAPE")
    print("=" * 50)
    print(df.shape)

    print("\n")

    print("=" * 50)
    print("COLUMN NAMES")
    print("=" * 50)
    print(df.columns.tolist())

    print("\n")

    print("=" * 50)
    print("DATA TYPES")
    print("=" * 50)
    print(df.dtypes)

    print("\n")

    print("=" * 50)
    print("MISSING VALUES")
    print("=" * 50)
    print(df.isnull().sum())

    print("\n")

    print("=" * 50)
    print("DUPLICATES")
    print("=" * 50)
    print(df.duplicated().sum())
```

```python
sample_data = {
    "NDVI": [0.75, 0.42, 0.15, 0.67, None],
    "NDBI": [0.20, 0.55, 0.80, 0.30, 0.65],
    "LST": [28.5, 35.2, 41.8, 30.4, 38.7],
    "PopulationDensity": [1200, 4500, 8000, 2000, 6000],
    "Humidity": [65, 55, 40, 70, 50]
}

df = pd.DataFrame(sample_data)

df
```

```python
audit_data(df)
```
