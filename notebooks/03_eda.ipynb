```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample_data = {
    "NDVI": [0.75, 0.42, 0.15, 0.67, 0.42],
    "NDBI": [0.20, 0.55, 0.80, 0.30, 0.65],
    "LST": [28.5, 35.2, 41.8, 30.4, 38.7],
    "PopulationDensity": [1200, 4500, 8000, 2000, 6000],
}
```

```python
df = pd.DataFrame(sample_data)

df
```

```python
print("Dataset Shape:")
print(df.shape)
```

```python
df.describe()
```

```python
df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.scatter(
    df["NDVI"],
    df["LST"]
)

plt.xlabel("NDVI")
plt.ylabel("LST")
plt.title("NDVI vs Land Surface Temperature")

plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.scatter(
    df["NDBI"],
    df["LST"]
)

plt.xlabel("NDBI")
plt.ylabel("LST")
plt.title("NDBI vs Land Surface Temperature")

plt.show()
```

```python
plt.figure(figsize=(6,4))

df["LST"].hist()

plt.title("Distribution of Land Surface Temperature")
plt.xlabel("LST")
plt.ylabel("Frequency")

plt.show()
```
