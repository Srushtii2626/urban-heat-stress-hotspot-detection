```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
importance = pd.DataFrame({
    "Feature": [
        "NDVI",
        "NDBI",
        "PopulationDensity",
        "Humidity",
        "WindSpeed"
    ],
    "Importance": [
        0.18,
        0.42,
        0.22,
        0.10,
        0.08
    ]
})

importance
```

```python
importance.sort_values(
    by="Importance",
    ascending=False
)
```

```python
importance.sort_values(
    by="Importance",
    ascending=True
).plot(
    x="Feature",
    y="Importance",
    kind="barh",
    figsize=(8,5)
)

plt.title("Urban Heat Drivers")

plt.show()
```

```python
top_driver = importance.sort_values(
    by="Importance",
    ascending=False
).iloc[0]

print("Top Heat Driver:")
print(top_driver)
```
