```python
import pandas as pd
```

```python
current_data = {
    "NDVI": [0.25],
    "NDBI": [0.70],
    "Predicted_LST": [41.0]
}

df = pd.DataFrame(current_data)

df
```

```python
print("Current Scenario")
df
```

```python
scenario_1 = df.copy()

scenario_1["NDVI"] = 0.45

scenario_1
```

```python
scenario_1["Predicted_LST"] = (
    scenario_1["Predicted_LST"] - 3
)

scenario_1
```

```python
print("""
Scenario:

Increase Green Cover

NDVI:
0.25 → 0.45

Estimated LST Reduction:
41°C → 38°C
""")
```

```python
scenario_2 = df.copy()

scenario_2["NDBI"] = 0.55

scenario_2["Predicted_LST"] = (
    scenario_2["Predicted_LST"] - 2
)

scenario_2
```

```python
print("""
Scenario:

Cool Roof Program

Built-up Heat Effect Reduced

Estimated Temperature Reduction:
41°C → 39°C
""")
```
